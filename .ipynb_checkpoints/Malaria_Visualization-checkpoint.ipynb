```python
deaths_df = deaths_df.rename(columns={'Deaths - Malaria - Sex: Both - Age: Age-standardized (Rate) (per 100,000 people)':'Age-standardized Deaths per 100,000', 'Entity':'Country'})

## Standardizing columns for deaths_df
```

```python
deaths_age_df.columns = [i.capitalize() for i in deaths_age_df.columns]
deaths_age_df = deaths_age_df.rename(columns={'Deaths':'Total Deaths', 'Entity':'Country'})
deaths_age_df ['Total Deaths'] = deaths_age_df ['Total Deaths'].round().astype(int)
deaths_age_df = deaths_age_df.drop('Unnamed: 0',axis=1)

## Standardizing columns for deaths_age_df
```

```python
incidence_df = incidence_df.rename(columns={'Incidence of malaria (per 1,000 population at risk) (per 1,000 population at risk)':'Incidence per 1,000 at risk', 'Entity':'Country'})

## Standardizing columns for incidence_df
```

```python
deaths_age_df1 = deaths_age_df.groupby(['Country','Year']).sum().reset_index()
deaths_age_df1.head(10)
```

```python
deaths_df = deaths_df.merge(deaths_age_df1,on=['Country','Year'])

## doing an inner join
```

```python
deaths_df
```

```python
def country_to_continent(country):
    
    """ This function converts country to continent. If country name not found, return NaN """
    
    try:
        country_code = pc.country_name_to_country_alpha2(country, cn_name_format="default")
        continent_code = pc.country_alpha2_to_continent_code(country_code)
        continent_name = pc.convert_continent_code_to_continent_name(continent_code)        
    except KeyError:
        continent_name = np.nan
    
    return continent_name
```

```python
def year_group(year):
    if year>=1990 and year<= 1994:
        return '90-94'
    if year>=1995 and year<= 1999:
        return '95-99'    
    if year>=2000 and year<= 2004:
        return '00-04'  
    if year>=2005 and year<= 2009:
        return '05-09'  
    if year>=2010 and year<= 2014:
        return '10-14'    
    if year>=2015 and year<= 2016:
        return '15-16'        
```

```python
deaths_df['Continent'] = deaths_df['Country'].apply(country_to_continent) 
deaths_df['Year Group'] = deaths_df['Year'].apply(year_group)
```

```python
group_total_death = deaths_df.groupby(['Continent','Year Group'])['Age-standardized Deaths per 100,000'].mean().unstack()
group_total_death.reset_index()
group_total_death = group_total_death.reindex(columns=["90-94","95-99","00-04","05-09","10-14","15-16"])
group_total_death.sort_values(by='90-94',ascending=False, inplace=True) 
group_total_death.drop('Europe',axis=0,inplace=True)
group_total_death

## creating pivot-table for heatmap
```

```python
from matplotlib.colors import LogNorm
fig = plt.figure(figsize=[8,6])

ax = sns.heatmap(group_total_death,cmap='Blues',linecolor='white',linewidths=1,annot = True,norm=LogNorm())
plt.xlabel('Year Range',fontsize=15)
plt.ylabel('Continent',fontsize=15)   
plt.title('Mean Deaths per 100,000 People',weight='bold',fontsize=15)
plt.savefig('heatmap.png',transparent=True, bbox_inches='tight')

## colour bar is log scaled
```

```python
deaths_age_df['Continent'] = deaths_age_df['Country'].apply(country_to_continent) 
len(deaths_age_df)
```

```python
deaths_age_df.dropna(axis=0,inplace=True)
len(deaths_age_df)
```

```python
group_age_death = deaths_age_df.groupby(['Year','Age_group'])['Total Deaths'].sum().reset_index()
group_age_death
```

```python
fig = plt.figure(figsize=[12,8])
ax = sns.lineplot(x='Year',y='Total Deaths',data=group_age_death,hue='Age_group',style='Age_group',markers=True)
plt.legend(loc="upper right", bbox_to_anchor=(1.0, 0.6),fontsize=12)
plt.xlabel('Year',weight='bold',fontsize=15)
plt.ylabel('Total Deaths from Malaria',weight='bold',fontsize=15)
plt.title('Time Series of Total Deaths',weight='bold',fontsize=18)
ax.set_yticklabels([str(int(label/1000))+'k' for label in ax.get_yticks()])
sns.despine()
plt.savefig('lineplot.png',transparent=True, bbox_inches='tight')

## Total deaths corresponds with true values from official results
```

```python
deaths_df.dropna(axis=0,inplace=True)
```

```python
deaths_df_2016 = deaths_df[deaths_df['Year']==2016]
deaths_df_2016['Log Total Deaths'] = np.log(deaths_df_2016['Total Deaths'])
deaths_df_2016['Log Age-standardized Deaths per 100,000'] = np.log(deaths_df_2016['Age-standardized Deaths per 100,000'])
```

```python
fig = plt.figure(figsize=[10,8])
ax = sns.regplot(x='Log Age-standardized Deaths per 100,000', y='Log Total Deaths',data= deaths_df_2016,fit_reg=False)
plt.xlabel('Log(Deaths per 100,000 people)',fontsize=15)
plt.ylabel('Log(Total Deaths)',fontsize=15)
plt.title('Relationship between Total Death and Death Rate (2016)',fontsize=15)
sns.despine()
plt.savefig('regplot.png',transparent=True, bbox_inches='tight')

## Log of total deaths follows a linear relation with the log of death rate
```
